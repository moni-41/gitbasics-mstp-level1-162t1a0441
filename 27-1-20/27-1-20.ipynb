```python
sns.set(style='darkgrid')
i=sns.load_dataset("iris")
i.head()
```

```python
i["species"].value_counts()
```

```python
sns.relplot(x="sepal_width",y="petal_width",data=i,hue="species",col="species")
```

```python
#box plot
f,axs=plt.subplots(1,2)
sns.boxplot(x="species",y="sepal_length",data=i,ax=axs[0])
```

```python
#box plot
f,axs=plt.subplots(1,2)
sns.boxplot(x="species",y="sepal_length",data=i,ax=axs[0])
sns.scatterplot(x="species",y="sepal_length",data=i,ax=axs[1])
```

```python
#box plot
f,axs=plt.subplots(1,2)
sns.boxplot(x="species",y="sepal_length",data=i,ax=axs[0])
sns.scatterplot(x="species",y="sepal_length",data=i,ax=axs[1])
```

```python
#box plot
f,axs=plt.subplots(1,2)
sns.boxplot(x="species",y="sepal_length",data=i,ax=axs[0])
sns.scatterplot(x="species",y="sepal_length",hue="petal_length",data=i,ax=axs[1])
```

```python
sns.boxplot(x="sepal_width",y="petal_width",data=i)
```

```python
sns.jointplot(x="sepal_width",y="petal_width",data=i,kind="kde")
```

```python
sns.jointplot(x="sepal_width",y="petal_width",data=i,kind="hex")
```

```python
#pair plot
sns.pairplot(data=i)
```

```python
sns.pairplot(data=i,hue="species",palette="husl")
```

```python
sns.pairplot(data=i,hue="species",markers=["x","s","d"])
```

```python
sns.pairplot(i,vars=["sepal_width","sepal_length"])
```

```python
sns.pairplot(i,diag_kind="kde",markers="+")
```
